# 1. Business Understanding
## 1.1 Industry Overview
Banking Industry

The banking industry plays a vital role in Indonesia's financial system by serving as a financial intermediary, collecting funds from the public (funding) and redistributing them in the form of loans or financing (lending). According to Law No. 10 of 1998 concerning Banking, a bank is defined as a business entity that collects funds from the public in the form of deposits and channels them back to the public through loans and/or other financing facilities to improve the standard of living of the community.

Through this intermediation function, banks primarily generate revenue from the difference between the interest earned on loans and the interest paid to depositors, commonly referred to as Net Interest Income (NII). Therefore, a bank's ability to attract and maintain customer deposits is essential for ensuring liquidity, supporting lending activities, and sustaining long-term business growth.

## 1.2 Business Context
Third-Party Funds (DPK) and Term Deposits

The primary source of funding for banks comes from Third-Party Funds (Dana Pihak Ketiga/DPK), which consist of public deposits in the form of current accounts, savings accounts, and term deposits. These funds serve as the primary capital source that enables banks to perform their intermediation function by providing loans to individuals and businesses.

One of the key components of DPK is the term deposit, a fixed-term deposit in which customers place their funds for a specified period in exchange for a higher interest rate than a regular savings account. Unlike savings accounts, which allow withdrawals at any time, term deposits provide a more stable funding source because the funds are generally accessible only upon maturity, as agreed between the bank and the customer.

This stability makes term deposits strategically important for banks. The funds collected can be used to support lending activities, maintain liquidity, and improve cash flow planning. A stable funding base also enables banks to better manage liquidity risk while supporting business expansion and long-term financing activities.

However, competition among banks to attract Third-Party Funds has become increasingly intense. Banks continuously develop various marketing strategies to acquire and retain customers, including telemarketing campaigns to promote term deposit products. Nevertheless, not all customers have the financial needs or interest to invest in term deposits. As a result, telemarketing campaigns are often directed toward customers with a low likelihood of subscribing, leading to higher operational costs, lower marketing productivity, and suboptimal campaign conversion rates.

##4.4.1 Age

Age is one of the fundamental demographic characteristics that may influence customers' financial needs and investment decisions. This analysis examines whether certain age groups demonstrate higher subscription rates for term deposits.

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[17,30,40,50,60,100],
    labels=[
        '18-30',
        '31-40',
        '41-50',
        '51-60',
        '60+'
    ]
)
```

```python
age_conversion = (
    df.groupby('age_group')['y']
      .apply(lambda x:(x=='yes').mean()*100)
      .round(2)
      .reset_index(name='Conversion Rate (%)')
)

age_conversion
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=age_conversion,
    x='age_group',
    y='Conversion Rate (%)'
)

ax = sns.barplot(
    data=age_conversion,
    x='age_group',
    y='Conversion Rate (%)'
)

for i, value in enumerate(age_conversion['Conversion Rate (%)']):
    ax.text(
        i,
        value + 0.5,
        f'{value:.1f}%',
        ha='center',
        fontsize=10
    )

plt.title('Conversion Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Conversion Rate (%)')

plt.show()
```

### Interpretation

The conversion rate varies across different age groups. Customers aged 60 years and above exhibit the highest conversion rate, indicating that this age segment is the most likely to subscribe to a term deposit.

In contrast, customers aged between 31–60 years show relatively lower conversion rates, while the 18–30 age group demonstrates a moderate conversion rate. These findings suggest that customer age is associated with differences in subscription behavior and may serve as an important segmentation variable in future marketing campaigns.

### Business Insight

Customers aged 60 years and above represent a promising target segment for future telemarketing campaigns due to their significantly higher conversion rate. This behavior may reflect a stronger preference for low-risk savings and investment products among older customers.

Rather than distributing marketing efforts evenly across all age groups, the bank can prioritize customers in this segment while continuing to evaluate other demographic and financial characteristics to improve campaign effectiveness.

```python
# Customer Distribution by Age Group
age_count = (
    df.groupby("age_group", observed=False)
      .size()
      .reindex(["18-30", "31-40", "41-50", "51-60", "60+"])
      .reset_index(name="Total Customers")
)

plt.figure(figsize=(9,5))

ax = sns.barplot(
    data=age_count,
    x="age_group",
    y="Total Customers",
    color="#4C78A8",
    edgecolor="black"
)

# Title
plt.title(
    "Customer Distribution by Age Group",
    fontsize=15,
    fontweight="bold",
    pad=15
)

# Axis labels
plt.xlabel("Age Group", fontsize=11)
plt.ylabel("Number of Customers", fontsize=11)

# Grid
plt.grid(axis="y", linestyle="--", alpha=0.3)
plt.grid(axis="x", visible=False)

# Value labels
for container in ax.containers:
    ax.bar_label(
        container,
        fmt="{:,.0f}",
        fontsize=10,
        padding=3
    )

# Remove top/right border
sns.despine()

plt.tight_layout()
plt.show()
```

**Interpretation**

The largest customer segment belongs to 31–40 years old, followed by 41–50 years old, while customers aged 60+ represent the smallest proportion of the customer base.

**Business Insight**

Although customers aged 60+ demonstrate the highest conversion rate, they account for a relatively small share of the customer population. Therefore, both conversion performance and segment size should be considered when prioritizing marketing campaigns.

##4.4.2 Job

Occupation may reflect customers' income stability, financial literacy, and investment preferences. This analysis aims to identify which occupational groups are more likely to subscribe to a term deposit.

```python
job_conversion = (
    df.groupby('job')['y']
      .apply(lambda x:(x=='yes').mean()*100)
      .sort_values(ascending=False)
      .reset_index(name='Conversion Rate (%)')
)

job_conversion
```

```python
job_summary = (
    df.groupby('job')
      .agg(
          Total_Customers=('y', 'count'),
          Subscribers=('y', lambda x: (x == 'yes').sum())
      )
)

job_summary['Conversion Rate (%)'] = (
    job_summary['Subscribers'] /
    job_summary['Total_Customers'] * 100
).round(2)

job_summary.sort_values(
    by='Conversion Rate (%)',
    ascending=False
)
```

```python
plt.figure(figsize=(10,6))

ax = sns.barplot(
    data=job_conversion,
    y='job',
    x='Conversion Rate (%)'
)

for i, value in enumerate(job_conversion['Conversion Rate (%)']):
    ax.text(
        value + 0.3,
        i,
        f'{value:.1f}%',
        va='center',
        fontsize=10
    )

plt.title('Conversion Rate by Occupation')
plt.xlabel('Conversion Rate (%)')
plt.ylabel('Occupation')

plt.show()
```

### Interpretation

The conversion rate varies considerably across occupational groups. Students demonstrate the highest conversion rate (31.4%), followed by retired customers (25.3%). These groups appear to be substantially more likely to subscribe to a term deposit compared to other occupations.

Conversely, blue-collar and services employees exhibit the lowest conversion rates, suggesting a lower likelihood of subscribing to the product through telemarketing campaigns.

### Business Insight

Occupation appears to be an important customer segmentation variable. Students and retired customers represent promising target segments for future telemarketing campaigns due to their relatively high conversion rates.

However, these findings should be interpreted alongside the size of each customer group. A high conversion rate does not necessarily imply that the group contributes the largest number of subscriptions if its population is relatively small. Therefore, occupation should be considered together with other customer characteristics when prioritizing campaign targets.

## **4.5.1 Contact Strategy** ##

### Hypothesis

H₀: Contact type has no significant relationship with term deposit subscription.

H₁: Contact type significantly influences the likelihood of customer subscription.

### Visualization

```python
contact_rate = (
    df.groupby('contact')['y']
      .apply(lambda x: (x=='yes').mean()*100)
      .reset_index(name='Conversion Rate')
)

plt.figure(figsize=(6,5))

ax = sns.barplot(
    data=contact_rate,
    x='contact',
    y='Conversion Rate',
    palette='Set2'
)

for i,v in enumerate(contact_rate['Conversion Rate']):
    plt.text(i,v+0.5,f'{v:.1f}%',ha='center')

plt.title('Conversion Rate by Contact Type')
plt.xlabel('Contact Type')
plt.ylabel('Conversion Rate (%)')

plt.show()
```

### Chi-Square Test

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df['contact'],df['y'])

chi2,p,dof,expected = chi2_contingency(contingency)

print("Chi-square :",round(chi2,3))
print("P-value :",p)
```

### Interpretation

Customers contacted via cellular phones achieve a substantially higher conversion rate than those contacted through traditional telephone calls. The Chi-Square test produces a p-value < 0.05, indicating that contact type has a statistically significant relationship with subscription decisions.

### Business Insight

The bank's telemarketing resources can be utilized more efficiently by prioritizing cellular-based communication over traditional telephone calls.

The higher conversion rate through cellular contact may be attributed to several factors, including:

greater customer accessibility,
increased likelihood of answering mobile calls,
more direct and personal communication,
changing customer communication preferences.

While telephone contact may still be appropriate for certain customer segments, allocating a larger proportion of campaign efforts to cellular communication is likely to improve overall campaign effectiveness and marketing return on investment (ROI).

This insight also suggests that communication strategy should be incorporated into future campaign planning alongside customer segmentation.

Although the illiterate education category recorded the highest conversion rate, this group consisted of only 18 customers in the dataset. Consequently, the result is highly sensitive to small changes in the number of subscribers and should not be interpreted as a representative customer segment for business decision-making. From a practical perspective, customers with university degrees represent a more reliable target segment because they combine relatively high conversion rates with a substantially larger customer population.

The analysis also shows that campaign execution plays an important role in telemarketing success. Customers contacted via cellular phones exhibited significantly higher conversion rates than those contacted through traditional telephone calls. In addition, customers with successful previous campaign outcomes or previous marketing interactions were considerably more likely to subscribe to a term deposit than customers without prior engagement. Campaign performance also varied across different months, suggesting that campaign timing should be considered when allocating marketing resources.

Overall, the analysis demonstrates that telemarketing effectiveness can be improved by combining better customer segmentation, more effective communication strategies, and the utilization of historical campaign information to prioritize customers with higher subscription potential.

## **5.2 Business Recommendation** ##

Based on the analytical findings, several recommendations are proposed to improve the effectiveness of future telemarketing campaigns.

**Recommendation 1: Prioritize High-Potential Customer Segments**

Bank Garuda should implement a customer prioritization strategy by focusing telemarketing campaigns on customer segments with historically higher conversion rates, particularly customers aged 60 years and above, students, retired customers, and customers with university degrees.

Success Metric:
- Increase the overall campaign conversion rate from 11.27% to at least 13% within the next 12 months.

Business Value:
- Prioritizing customers with higher subscription potential is expected to improve marketing efficiency while reducing unnecessary outbound calls.

**Recommendation 2: Prioritize Cellular Communication**

Whenever customer contact information is available, telemarketing agents should prioritize cellular communication over traditional telephone calls, as cellular contact consistently achieves higher conversion rates.

Furthermore, a recent banking case study conducted at PT KB Bukopin Jember Branch found that a hybrid interaction strategy, combining digital communication with face-to-face personal interaction, was the most effective approach for building customer trust and converting leads into banking customers.

Therefore, instead of replacing telemarketing entirely, the bank should use predictive analytics to identify customers with high subscription potential and assign those customers to Relationship Managers, while lower-priority customers can continue to be served through telemarketing or digital marketing channels.

Success Metric:

- Increase the proportion of telemarketing campaigns conducted through cellular communication to at least 80% within six months.
- Monitor monthly conversion rates to evaluate performance improvements.

Business Value:
- Optimizing communication channels can improve customer reach and increase campaign effectiveness without requiring significant additional investment.



**Recommendation 3: Implement Customer Priority Scoring Based on Previous Campaign History**

Develop a simple customer priority scoring system that incorporates previous campaign outcomes and previous contact history when selecting customers for telemarketing campaigns.

Customers with successful previous campaign outcomes or positive engagement history should receive higher priority during campaign planning.

Success Metric:

- Implement the scoring framework within three months.
- Evaluate campaign performance after each quarterly campaign and compare conversion rates between prioritized and non-prioritized customer groups.

Business Value:
- Leveraging historical customer interactions enables the bank to focus marketing resources on customers with the highest probability of conversion.

**Recommendation 4: Optimize Campaign Planning Based on Historical Performance**

Campaign schedules should be reviewed annually using historical campaign data to identify periods with consistently higher customer responsiveness.

Marketing resources should be allocated more strategically during historically stronger campaign periods while considering campaign volume and operational capacity.

Success Metric:

- Conduct campaign performance reviews at least once per year.
- Increase campaign conversion rates during high-performing periods by 10% compared with previous campaigns.

Business Value:
- Strategic campaign scheduling can improve marketing efficiency and maximize resource utilization.

**Recommendation 5: Adopt a Hybrid Marketing Strategy**

Rather than relying exclusively on outbound telemarketing, Bank Garuda should combine telemarketing with Relationship Manager (RM) engagement for customers identified as having high subscription potential.

Relationship Managers can provide personalized financial consultations, explain the benefits of term deposits, and build stronger customer relationships before investment decisions are made.

Success Metric:

- Compare conversion rates between customers served through telemarketing only and those receiving RM assistance.

Business Value:
- Combining personalized relationship management with telemarketing can improve customer trust and increase subscription rates for higher-value customers.

**Recommendation 6: Develop a Data-Driven Customer Segmentation Framework**

Develop an integrated customer segmentation framework that combines:
- demographic characteristics,
- previous campaign outcomes,
- previous contact history,
- campaign characteristics,

Success Metric:
